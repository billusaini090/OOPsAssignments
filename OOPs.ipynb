#**Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.**



 **Class**

A class is a blueprint or template used to create objects.
It defines data members (variables) and member functions (methods) but does not occupy memory until an object is created.

Example: A class Car may define properties like colour, model, speed and behaviours like start(), stop().

**Object**

An object is a real-world entity created from a class.
It is an instance of the class and occupies memory.
Objects can access the data and functions defined in the class.

Example:
If Car is a class, then

myCar

yourCar
are objects of that class with different values.

```python
# Example--

class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def show_info(self):
        print("Brand:", self.brand, ", Speed:", self.speed)


# Creating an object of Car class
c1 = Car("Honda", 120)

# Calling object method
c1.show_info()
```

#**Q3. Explain why the __init__() function is used. Give a suitable example.**

The __init__() function in Python is a constructor that automatically runs when an object is created.
It is used to initialize the object’s data (variables).

```python
#Example--
class Student:
    def __init__(self, name, age):   # constructor
        self.name = name
        self.age = age

    def show(self):
        print("Name:", self.name, "Age:", self.age)


# Creating object
s1 = Student("Rohan", 18)

s1.show()

```

#**Q4. Why self is used in OOPs?**


---


In Python OOPs, self is used to refer to the current object, allowing the class methods to access the object's variables and functions.

One-line explanation:

#**Q5. What is inheritance? Give an example for each type of inheritance**.

**Inheritance is an OOPs feature in which one class (child) acquires the properties and methods of another class (parent).
It helps in code reusability and method extension.**



#Types of Inheritance with Examples (Python)
1. Single Inheritance

One child class inherits from one parent class.

```python
#Example --Single Inheritance

class A:
    def showA(self):
        print("This is Class A")

class B(A):
    def showB(self):
        print("This is Class B")

obj = B()
obj.showA()
obj.showB()
```

2. Multilevel Inheritance

A → B → C (chain of inheritance

```python
#Example -- Multilevel Inheritance.

class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(B):
    def showC(self):
        print("C")

obj = C()
obj.showA()
obj.showB()
obj.showC()
```

3. Multiple Inheritance

A child class inherits from more than one parent class.

```python
#Example-- Multiple Inheritance



 class A:
    def showA(self):
        print("A")

class B:
    def showB(self):
        print("B")

class C(A, B):
    pass

obj = C()
obj.showA()
obj.showB()
```

4. Hierarchical Inheritance

One parent class → multiple child classes.

```python
#Example --Hirarchical Inheritance.

class A:
    def showA(self):
        print("Parent A")

class B(A):
    pass

class C(A):
    pass

obj1 = B()
obj2 = C()
obj1.showA()
obj2.showA()
```


5. Hybrid Inheritance

Combination of two or more types (e.g., multiple + hierarchical).

```python
#Example -- Hybrid Inheritance

class A:
    def showA(self):
        print("A")

class B(A):
    pass

class C(A):
    pass

class D(B, C):   # multiple + hierarchical = hybrid
    pass

obj = D()
obj.showA()
```
